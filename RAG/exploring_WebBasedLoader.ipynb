install required libray

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.document_loaders import RecursiveUrlLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain_community.vectorstores import FAISS
import streamlit as st
```

using webbaseloader without header

```python
loader_content=WebBaseLoader("https://www.uoh.edu.pk/")
loaded_document=loader_content.load()
splitted_text=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
text_document=splitted_text.split_documents(loaded_document)

```

```python
text_document[:1]
```

using webbaseloader with header

```python
loader_content=WebBaseLoader(web_path="https://www.uoh.edu.pk",header_template={"User-Agent":"Mozilla/5.0"})
loaded_document=loader_content.load()
splitted_text=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
text_document=splitted_text.split_documents(loaded_document)

```

```python
text_document[:1]
```

use recursiveurlloader to load all the content of uoh.
use header to getting data

```python
loaded_content=RecursiveUrlLoader(url="https://www.uoh.edu.pk",
                                   max_depth=2,
                                   headers={"User-Agent":"Mozilla/5.0"})
loaded_document=loaded_content.load()
text_document=splitted_text.split_documents(loaded_document)
```
