Importing Necessry Libraries to get the data From university of Haripur

```python
from langchain_community.document_loaders import RecursiveUrlLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
loaded=RecursiveUrlLoader("https://uoh.edu.pk")
load_documents=loaded.load()
spliter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
split_text=spliter.split_documents(load_documents)
```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings
embeddins=HuggingFaceBgeEmbeddings(model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2",
                                  model_kwargs={"device":"cpu"},
                                   encode_kwargs={"normalize_encode":False}
                                   )
```

```python
import getpass
import os
import time

from pinecone import Pinecone, ServerlessSpec

if not os.getenv("PINECONE_API_KEY"):
    os.environ["PINECONE_API_KEY"] = getpass.getpass("Enter your api key:  ")

pinecone_api_key = os.environ.get("PINECONE_API_KEY")

pc = Pinecone(api_key=pinecone_api_key)
```

```python
import time

index_name = "agent-ai"  # change if desired

existing_indexes = [index_info["name"] for index_info in pc.list_indexes()]

if index_name not in existing_indexes:
    pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1"),
    )
    while not pc.describe_index(index_name).status["ready"]:
        time.sleep(1)

index = pc.Index(index_name)
```

```python
from langchain_pinecone import PineconeVectorStore

# Store the embeddings into Pinecone
vector_store = PineconeVectorStore.from_documents(
    documents=split_text,  # The split documents from WebBaseLoader
    embedding=embeddins,  # The embeddings model
    index_name=index_name,  # The Pinecone index name
)

print("Documents successfully stored in Pinecone.")
```

```python
#vector_store.similarity_search("today Head lines")
query = "Faculty of Biological & Biomedical Sciences"
results = vector_store.similarity_search(
    query,
    k=1
)
for res in results:
    print(f"* {res.page_content} [{res.metadata}]")
```

```python
uoh_retriever = vector_store.as_retriever()
```

```python
from langchain.tools.retriever import create_retriever_tool
uoh_tool=create_retriever_tool(uoh_retriever,"Search In University of Haripur","if you get any prompt regarding university of haripur or uoh. Your must be use this tool to response particular answer")
```

```python
api_wrapper=ArxivAPIWrapper(top_k_results=1,doc_content_chars_max=200)
arxiv_tool=ArxivQueryRun(api_wrapper=api_wrapper)
```

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.tools.retriever import create_retriever_tool
web_doc=WebBaseLoader("https://www.bbc.com/urdu",header_template={"User-Agent":"Mozilla/5.0"})

```

```python
load_doc=web_doc.load()
split_doc=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
split_content=split_doc.split_documents(load_doc)
```

```python
embedding=HuggingFaceBgeEmbeddings(
    model_name="FacebookAI/xlm-roberta-base",# for urdu text embeddings
    model_kwargs={"device":"cpu"},
    encode_kwargs={"normalize_embedding":False}
)
```

```python
chroma_db=Chroma.from_documents(split_content,embedding=embedding)
urdupoint_retriever=chroma_db.as_retriever
```

```python
urdupoint_tool=create_retriever_tool(urdupoint_retriever,"Search in BBC Urdu","If you recieve anything related to Urdu News or urdu text you must use this tool")
```

```python
tools=[uoh_tool,arxiv_tool,urdupoint_tool]
```

```python
from langchain import hub
# Get the prompt to use - you can modify this!
prompt = hub.pull("hwchase17/openai-functions-agent")
prompt.messages
```

```python
from langchain.agents import create_openai_tools_agent
agents=create_openai_tools_agent(llm=llm,prompt=prompt,tools=tools)
```

```python
from langchain.agents import AgentExecutor
agent_executer=AgentExecutor(agent=agents,tools=tools,verbose=True)
```

```python
result=agent_executer.invoke({"input":"what is BBC Urdu"})
```
